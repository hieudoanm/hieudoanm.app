```python
import pandas
import numpy
```

```python
players_df = pandas.read_csv("./players.csv")
players_df
```

```python
players_df["world_titles"] = (
    players_df["world_champion"]
    + players_df["world_rapid"]
    + players_df["world_blitz"]
    + players_df["world_fisher"]
    + players_df["world_cup"]
)
players_df = players_df.sort_values(by=["world_titles", "peak_rating"], ascending=[False, True])
players_df
```

```python
def get_generation(year: int) -> str:
    if year >= 1901 and year <= 1927:
        return "The Greatest Generation"
    if year >= 1928 and year <= 1945:
        return "The Silent Generation"
    if year >= 1946 and year <= 1964:
        return "Baby Boom Generation"
    if year >= 1965 and year <= 1980:
        return "Generation X"
    if year >= 1981 and year <= 1996:
        return "Generation Y"
    if year >= 1997 and year <= 2010:
        return "Generation Z"
    return ""
```

```python
players_df["generation"] = players_df["birth_year"].apply(get_generation)
players_df["achieved_age"] = players_df["peak_year"] - players_df["birth_year"]
players_df["achieved_age_range"] = numpy.floor(players_df["achieved_age"] / 10) * 10
players_df["achieved_decade"] = numpy.floor(players_df["peak_year"] / 10) * 10
players_df = players_df.sort_values(by=["achieved_age", "peak_rating"], ascending=[True, False])
players_df
```
