```python
import dotenv
import os
import pandas
import requests
```

```python
dotenv.load_dotenv()

URL = "https://api.football-data.org/v4"

API_KEY_FOOTBALL_DATA = os.getenv("API_KEY_FOOTBALL_DATA")
```

```python
response = requests.get(f"{URL}/competitions")
data = response.json()
competitions = data.get("competitions", [])
competitions_data_frame = pandas.DataFrame(competitions)
competitions_data_frame.to_csv("./competitions.csv", index=False)
competitions_data_frame
```

```python
response = requests.get(f"{URL}/competitions/2021", headers={
    "X-AUTH-TOKEN": API_KEY_FOOTBALL_DATA
})
data = response.json()
data
```

```python
response = requests.get(f"{URL}/competitions/2021/teams", headers={
    "X-AUTH-TOKEN": API_KEY_FOOTBALL_DATA
})
data = response.json()
teams = data.get("teams", [])
teams_data_frame = pandas.DataFrame(teams)
teams_data_frame.to_csv("./teams.csv", index=False)
teams_data_frame
```

```python
response = requests.get(f"{URL}/teams/64/matches", headers={
    "X-AUTH-TOKEN": API_KEY_FOOTBALL_DATA
})
data = response.json()
matches = data.get("matches", [])
matches_data_frame = pandas.DataFrame(matches)
matches_data_frame.to_csv("./matches.csv", index=False)
matches_data_frame
```
