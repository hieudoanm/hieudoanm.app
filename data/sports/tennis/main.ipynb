```python
import pandas
```

```python
champions_data_frame = pandas.read_csv("./champions.csv")
champions_data_frame
```

```python
competitions = champions_data_frame["competition"].unique()
competitions.sort()
competitions
```

## Olympics

```python
olympics_data_frame = champions_data_frame[champions_data_frame["competition"] == "Olympics"]
olympics_data_frame
```

```python
olympics_champions = olympics_data_frame["champion"].value_counts()
olympics_champions
```
