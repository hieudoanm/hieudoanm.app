```python
import requests
```

```python
# BASE_URL = "https://diamondchessapi.vercel.app/api/chess"
BASE_URL = "http://localhost:8080/api/chess"
```

```python
USERNAME = "hikaru"
```

```python
response = requests.get(f'{BASE_URL}/player/{USERNAME}')
player = response.json()
player
```

```python
archives = player.get("archives")
archivesYearMonth = []
for archive in archives:
    [year, month] = archive.split("/")
    archivesYearMonth.append({
        "year": year,
        "month": month
    })
archivesYearMonth
```

```python
for item in archivesYearMonth:
    month = item.get("month")
    year = item.get("year")
    data = {
        "month": month,
        "year": year
    }
    response = requests.post(f'{BASE_URL}/player/{USERNAME}/games', json=data)
    data = response.json()
    print(year, month, data.get("synced"), data.get("total"))
```
