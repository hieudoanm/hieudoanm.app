```python
import requests
from tqdm.notebook import tqdm as tqdm_notebook
```

```python
BASE_URL = "https://diamondchessapi.vercel.app/api/chess"
```

```python
response = requests.get(f'{BASE_URL}/openings')
data = response.json()
openings = data.get("openings")
openings
```

```python
for opening in tqdm_notebook(openings):
    fen = opening.get("fen", "")
    response = requests.get(f'{BASE_URL}/position?fen={fen}')
    print(response.json())
```
