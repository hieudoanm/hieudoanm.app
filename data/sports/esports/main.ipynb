```python
import pandas as pd
```

```python
champions_data_frame = pd.read_csv("./champions.csv")
champions_data_frame = champions_data_frame.sort_values(by=['sport', 'competition', 'year'])
champions_data_frame.to_csv("./champions.csv", index=False)
champions_data_frame
```

## Worlds

```python
worlds_data_frame = champions_data_frame[champions_data_frame['competition'] == "Worlds"]
worlds_data_frame
```

```python
champions = worlds_data_frame[['champion', 'champion_country']].value_counts()
champions
```

```python
runners_up = worlds_data_frame[['runner_up', 'runner_up_country']].value_counts()
runners_up
```

```python
finals = worlds_data_frame[['champion', 'runner_up']].value_counts()
finals
```
