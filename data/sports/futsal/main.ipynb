```python
import pandas
```

```python
champions_data_frame = pandas.read_csv("./champions.csv")
champions_data_frame = champions_data_frame.sort_values(by=['sport', 'gender', 'competition', 'year'])
champions_data_frame
```

```python
competitions = champions_data_frame["competition"].unique()
competitions
```

## FIFA Futsal World Cup

```python
world_cup_data_frame = champions_data_frame[champions_data_frame["competition"] == "FIFA Futsal World Cup"]
world_cup_data_frame
```

```python
hosts = world_cup_data_frame['host'].value_counts()
hosts
```

```python
champions = world_cup_data_frame['champion'].value_counts()
champions
```

```python
runners_up = world_cup_data_frame['runner_up'].value_counts()
runners_up
```

```python
third_places = world_cup_data_frame['third_place'].value_counts()
third_places
```

```python
fourth_places = world_cup_data_frame['fourth_place'].value_counts()
fourth_places
```
