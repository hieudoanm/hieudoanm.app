```python
import pandas as pd
```

```python
secretaries_data_frame = pd.read_csv("./secretaries.csv")
sorted_secretaries_data_frame = secretaries_data_frame.sort_values(by=["department_no", "department", "head_no"])
sorted_secretaries_data_frame['department_no'] = sorted_secretaries_data_frame['department_no'].astype(int)
sorted_secretaries_data_frame['head_no'] = sorted_secretaries_data_frame['head_no'].astype(int)
sorted_secretaries_data_frame.to_csv("./secretaries.csv", index=False)
sorted_secretaries_data_frame
```

## Departments

```python
department_data_frame = sorted_secretaries_data_frame[sorted_secretaries_data_frame["department_no"] != 0]
department_data_frame = department_data_frame[["department_no", "department", "head"]]
department_data_frame = department_data_frame.drop_duplicates(["department_no", "department", "head"])
department_data_frame
```

## Secretaries of State

```python
state_data_frame = sorted_secretaries_data_frame[sorted_secretaries_data_frame["department"] == "Department of State"]
state_data_frame
```

```python
foreign_borned_state_data_frame = state_data_frame[state_data_frame["birth_country"] != "United States"]
foreign_borned_state_data_frame
```
