# Presidents

```python
import matplotlib.pyplot as plt
import pandas
```

```python
presidents_data_frame = pandas.read_csv("./presidents.csv")
presidents = presidents_data_frame.to_dict("records")
presidents_data_frame
```

## Ranking by Experiences

```python
presidents_rankings = []

for president in presidents:
    number = president.get("number")
    name = president.get("name")
    vice_point = 1 if president.get("former_vice") != 'none' else 0
    governor_point = 1 if president.get("former_governor") != 'none' else 0
    senator_point = 1 if president.get("former_senator") != 'none' else 0
    speaker_point = 1 if president.get("former_speaker") != 'none' else 0
    house_point = 1 if president.get("former_house") != 'none' else 0
    secretary_point = 1 if president.get("former_secretary") != 'none' else 0
    point = vice_point + governor_point + senator_point + speaker_point + house_point + secretary_point
    presidents_rankings.append({
        "number": number,
        "name": name,
        "point": point
    })

presidents_rankings_data_frame = pandas.DataFrame(presidents_rankings)
sorted_presidents_rankings_data_frame = presidents_rankings_data_frame.sort_values(by = ["point", "number"], ascending = [False, True])
sorted_presidents_rankings_data_frame
```
