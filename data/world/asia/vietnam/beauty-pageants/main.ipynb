# Miss Vietnam

```python
import pandas
```

```python
miss_vietnam_data_frame = pandas.read_csv("./miss-vietnam.csv")
miss_vietnam_data_frame["winner_age"] = (miss_vietnam_data_frame["year"] - miss_vietnam_data_frame["winner_birth_year"])
miss_vietnam_data_frame["first_runner_up_age"] = (miss_vietnam_data_frame["year"] - miss_vietnam_data_frame["first_runner_up_birth_year"])
miss_vietnam_data_frame["second_runner_up_age"] = (miss_vietnam_data_frame["year"] - miss_vietnam_data_frame["second_runner_up_birth_year"])
miss_vietnam_data_frame
```

## Winner

```python
miss_vietname_winner_average_age = round(miss_vietnam_data_frame["winner_age"].mean(), 2)
miss_vietname_winner_average_age
```

```python
miss_vietname_winner_max_age = miss_vietnam_data_frame["winner_age"].max()
miss_vietname_winner_max_age
```

```python
miss_vietname_winner_min_age = miss_vietnam_data_frame["winner_age"].min()
miss_vietname_winner_min_age
```

```python
miss_vietnam_winner_cities = miss_vietnam_data_frame["winner_city"].value_counts()
miss_vietnam_winner_cities
```

## First Runner Up

```python
miss_vietname_first_runner_up_average_age = round(miss_vietnam_data_frame["first_runner_up_age"].mean(), 2)
miss_vietname_first_runner_up_average_age
```

```python
miss_vietnam_first_runner_up_cities = miss_vietnam_data_frame["first_runner_up_city"].value_counts()
miss_vietnam_first_runner_up_cities
```

## Second Runner Up

```python
miss_vietname_second_runner_up_average_age = round(miss_vietnam_data_frame["second_runner_up_age"].mean(), 2)
miss_vietname_second_runner_up_average_age
```

```python
miss_vietnam_second_runner_up_cities = miss_vietnam_data_frame["second_runner_up_city"].value_counts()
miss_vietnam_second_runner_up_cities
```
