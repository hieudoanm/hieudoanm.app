```python
import pandas
import requests
```

```python
response = requests.get("https://restcountries.com/v3.1/all")
countries : list[dict] = response.json()
countries
```

```python
languages = {}
for country in countries:
    languages_per_country : dict = country.get("languages", {})
    for language in languages_per_country:
        languages[language] = languages_per_country[language]

languages_list = []
for language in languages:
    languages_list.append({
        "code": language,
        "name": languages[language]
		})

languages_list
```

```python
languages_data_frame = pandas.DataFrame(languages_list)
sorted_languages_data_frame = languages_data_frame.sort_values(by=["code"])
sorted_languages_data_frame.to_csv("./languages.csv", index=False)
sorted_languages_data_frame
```
