# Passport

```python
from bs4 import BeautifulSoup
import pandas
import requests
from tqdm.notebook import tqdm as tdqm_notebook
```

```python
response = requests.get("https://restcountries.com/v3/all")
countries = response.json()
```

```python
for country in tdqm_notebook(countries):
  independent = country.get("independent", False)
  if not independent:
    continue
  countryCommonName = country.get("name", {}).get("common", "")
  countryPath = "-".join(countryCommonName.split(" ")).lower()
  try:
    print(countryPath)
    URL = f"https://www.passportindex.org/passport/{countryPath}"
    response = requests.get(URL)
    html = response.text
    soup = BeautifulSoup(html, "html.parser")
    rows = soup.find_all("table")[0].find("tbody").find_all("tr")
    rowsList = list(rows)
    visas = []
    for row in rows:
      cells = row.find_all("td")
      cellsList = list(cells)
      countryText = cellsList[0].getText().strip()
      requirementText = cellsList[1].getText().strip()
      visa = {
        "country": countryText,
        "requirement": requirementText
      }
      visas.append(visa)
    visas_data_frame = pandas.DataFrame(visas)
    sorted_visas_data_frame = visas_data_frame.sort_values(by = ["requirement", "country"])
    sorted_visas_data_frame.to_csv(f"./csv/{countryPath}.csv", index=False, header=True)
  except:
    print(countryPath, "error")
```
