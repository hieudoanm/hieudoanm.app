```python
from bs4 import BeautifulSoup
import pandas
import requests
from tqdm.notebook import tqdm as tdqm_notebook
```

## Countries

```python
COUNTRIES_URL = "https://restcountries.com/v3/all"

response = requests.get(COUNTRIES_URL)
countries = response.json()

countries_data_frame = pandas.DataFrame(countries)
countries_data_frame.to_csv("./countries/all.csv", index=False, header=True)
countries_data_frame
```

## United Nation Members

```python
united_nation_countries_data_frame = countries_data_frame[countries_data_frame["unMember"]]
united_nation_countries_data_frame.to_csv("./countries/united-nation-members.csv", index=False, header=True)
united_nation_countries_data_frame
```

## Regions

```python
region_series = countries_data_frame['region'].value_counts(ascending=True)
region_series
```

```python
for item in region_series.items():
    region = item[0]
    countries_by_region_data_frame = countries_data_frame[countries_data_frame["region"] == region]
    countries_by_region_data_frame.to_csv(f"./regions/{region.lower()}.csv", index=False, header=True)
```
