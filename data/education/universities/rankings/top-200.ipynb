# Top 200

```python
from IPython.display import HTML
from functools import reduce
import json
import pandas
```

```python
qs_rankings_data_frame = pandas.read_csv("./csv/qs.csv")
shanghai_rankings_data_frame = pandas.read_csv("./csv/shanghai.csv")
times_rankings_data_frame = pandas.read_csv("./csv/times.csv")
```

```python
top_200_qs_rankings_data_frame : pandas.DataFrame = qs_rankings_data_frame.head(200)
top_200_qs_rankings_data_frame.to_csv("./csv/top-200-qs.csv", index=False)
top_200_qs_universities : list[str] = top_200_qs_rankings_data_frame['university'].unique()
top_200_qs_countries : list[str] = top_200_qs_rankings_data_frame['country'].unique()
top_200_qs_rankings_data_frame
```

```python
top_200_shanghai_rankings_data_frame : pandas.DataFrame = shanghai_rankings_data_frame.head(200)
top_200_shanghai_rankings_data_frame.to_csv("./csv/top-200-shanghai.csv", index=False)
top_200_shanghai_universities : list[str] = top_200_shanghai_rankings_data_frame['university'].unique()
top_200_shanghai_countries : list[str] = top_200_shanghai_rankings_data_frame['country'].unique()
top_200_shanghai_rankings_data_frame
```

```python
top_200_times_rankings_data_frame : pandas.DataFrame = times_rankings_data_frame.head(200)
top_200_times_rankings_data_frame.to_csv("./csv/top-200-times.csv", index=False)
top_200_times_universities : list[str] = top_200_times_rankings_data_frame['university'].unique()
top_200_times_countries : list[str] = top_200_times_rankings_data_frame['country'].unique()
top_200_times_rankings_data_frame
```

```python
# -*- coding: utf-8 -*-

top_200_universities : list[str] = list(
    set([
        *top_200_qs_universities, 
        *top_200_shanghai_universities,
        *top_200_times_universities
    ])
)
top_200_universities.sort()
# Write to File
top_200_universities_file = open('./json/top-200-universities.json', 'w', encoding="utf-8")
json.dump(top_200_universities, top_200_universities_file, indent=2, ensure_ascii=False)
top_200_universities_file.close()

len(top_200_universities)
```

```python
# -*- coding: utf-8 -*-

top_200_countries : list[str] = list(
    set([
        *top_200_times_countries,
        *top_200_shanghai_countries,
        *top_200_times_countries
    ])
)
top_200_countries.sort()
# Write to File
top_200_countries_file = open('./json/top-200-countries.json', 'w', encoding="utf-8")
json.dump(top_200_countries, top_200_countries_file, indent=2, ensure_ascii=False)
top_200_countries_file.close()

len(top_200_countries)
```

```python
top_200_data_frames = [
    top_200_qs_rankings_data_frame,
    top_200_shanghai_rankings_data_frame,
    top_200_times_rankings_data_frame
]
on = ['university', 'country']
top_200_data_frame = reduce(lambda left, right: pandas.merge(left, right, on=on, how='outer'), top_200_data_frames)

top_200_data_frame.fillna(101, inplace=True)
top_200_data_frame['rank_average'] = round(top_200_data_frame[['rank_qs', 'rank_shanghai', 'rank_times']].mean(axis=1), 2)

sorted_top_200_data_frame = top_200_data_frame.sort_values('rank_average', ascending=True)
sorted_top_200_data_frame = sorted_top_200_data_frame.reset_index(drop=True)
sorted_top_200_data_frame.to_csv("./csv/top-200-all.csv", index=False)
sorted_top_200_data_frame
```

```python
filtered_top_200_data_frame : pandas.DataFrame = sorted_top_200_data_frame[
    (sorted_top_200_data_frame["rank_qs"] < 101) & 
    (sorted_top_200_data_frame["rank_shanghai"] < 101) & 
    (sorted_top_200_data_frame["rank_times"] < 101)
]
filtered_top_200_data_frame.to_csv("./csv/top-200-filtered.csv", index=False)
filtered_top_200_data_frame
```

```python
sorted_filtered_top_200_data_frame = filtered_top_200_data_frame.sort_values(by=['country', 'rank_average'], ascending=True)
sorted_filtered_top_200_data_frame.to_csv("./csv/top-200-filtered-countries.csv", index=False)
sorted_filtered_top_200_data_frame
```

```python
filtered_out_top_200_data_frame : pandas.DataFrame = sorted_top_200_data_frame[
    (sorted_top_200_data_frame["rank_qs"] == 101) | (sorted_top_200_data_frame["rank_shanghai"] == 101) | (sorted_top_200_data_frame["rank_times"] == 101)
]
HTML(filtered_out_top_200_data_frame.to_html(index=False))
```
