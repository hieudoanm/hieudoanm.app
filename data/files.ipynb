```python
import pandas as pd
import os

pd.set_option('display.max_rows', None)
```

```python
def get_paths(directory : str, extension : str):
    csv_paths = []
    for root, _, files in os.walk(directory):
        for file in files:
            if file.endswith(extension):
                csv_paths.append(os.path.join(root, file))
    return csv_paths
```

```python
def get_num_lines(directory : str):
    csv_file = open(directory, 'r')
    num_lines = sum(1 for _ in csv_file)
    return num_lines
```

```python
def get_files(extension : str):
    # Replace 'your_directory_path' with the path of the directory you want to search
    directory_path = '.'
    csv_paths = get_paths(directory_path, extension)
    print(len(csv_paths))
    files = []
    for path in csv_paths:
        num_lines = get_num_lines(path)
        if num_lines < 100:
            print(path, num_lines)
        files.append({
            "file_path": path,
            "num_lines": num_lines
        })
    files_data_frame = pd.DataFrame(files)
    files_data_frame = files_data_frame.sort_values(by = ["num_lines"])
    files_data_frame = files_data_frame.reset_index()
    return files_data_frame
```
