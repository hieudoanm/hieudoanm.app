# Top-level Domains

```python
import requests
from tqdm.notebook import tqdm as tqdm_notebook
```

```python
response = requests.get("https://data.iana.org/TLD/tlds-alpha-by-domain.txt")
tlds : list[str] = []

if response.status_code == 200:
    tlds_text = response.text.lower()
    tlds = tlds_text.split("\n")
    tlds_file = open("./tlds.txt", "w")
    tlds_file.write(tlds_text.lower())
    tlds_file.close()

tlds : list[str] = list(filter(lambda item: item != "", tlds))
tlds.pop(0)
tlds
```

```python
when = "2023-05-20T05:34:00.021Z"
encrypted = "8cfdb189e722969be89107bfea58bdbdaeb02d0930fd96b8"

noun_tlds = []
verb_tlds = []

def write_txt(file_path : str, content : str):
    file_txt = open(file_path, "w")
    file_txt.write(content)
    file_txt.close()

for tld in tqdm_notebook(tlds):
    url = f"https://www.wordsapi.com/mashape/words/{tld}?when={when}&encrypted={encrypted}"
    response = requests.get(url)
    status_code = response.status_code
    if response.status_code == 200:
        data = response.json()
        word = data.get("word", "")
        results = data.get("results", [])
        if word == tld and len(results) > 0:
            part_of_speeches = list(set(list(map(lambda item: item.get("partOfSpeech", ""), results))))
            if "noun" in part_of_speeches:
                noun_tlds.append(word)
                write_txt("./noun_tlds.txt", "\n".join(noun_tlds))
            if "verb" in part_of_speeches:
                verb_tlds.append(word)
                write_txt("./verb_tlds.txt", "\n".join(verb_tlds))
    else:
        print(tld, status_code)
```
