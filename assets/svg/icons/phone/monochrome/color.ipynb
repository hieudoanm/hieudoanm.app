```python
import os
from pathlib import Path

def get_svg_files(directory):
    svg_files = []
    for root, _, files in os.walk(directory):
        svg_files_directory = [f"{root}/{file}" for file in files if file.lower().endswith('.svg')]
        svg_files = svg_files + svg_files_directory
    return svg_files

if __name__ == "__main__":
    # directory = input("Enter the directory path to search: ").strip()
    directory = './black'
    black_color = "rgb(27,27,29)"
    white_color = "rgb(255,255,255)"
    dark_red_color = "rgb(149,6,6)"
    bordeaux_red_color = "rgb(90,0,44)"
    supreme_red_color = "rgb(237,29,36)"
    gold_yellow_color = "rgb(239,191,4)"
    dark_green_color = "rgb(6,64,43)"
    emerald_green_color = "rgb(0,103,79)"
    if os.path.isdir(directory):
        black_svg_files = get_svg_files(directory)
        print(f"Total number of .svg files: {len(black_svg_files)}")
        for black_svg_file in black_svg_files:
            black_svg_io = open(black_svg_file, 'r')
            black_svg_string = black_svg_io.read()
            black_count = black_svg_string.count(black_color)
            if black_count == 1:
                colored_svg_string = black_svg_string.replace(black_color, dark_green_color)
                colored_svg_file : str = black_svg_file.replace("black", "green/dark")
                colored_svg_path = Path("/".join(colored_svg_file.split("/")[:-1]))
                colored_svg_path.mkdir(parents=True, exist_ok=True)
                colored_svg_filename = Path(colored_svg_file)
                colored_svg_filename.touch(exist_ok=True)
                colored_svg_io = open(colored_svg_file, "w+")
                colored_svg_io.write(colored_svg_string)
                colored_svg_io.close()
                print(black_svg_file, black_count)
    else:
        print("The provided path is not a valid directory.")
```
