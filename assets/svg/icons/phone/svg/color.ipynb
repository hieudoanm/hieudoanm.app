```python
import os
from pathlib import Path

def get_svg_files(directory):
    svg_files = []
    for root, _, files in os.walk(directory):
        svg_files_directory = [f"{root}/{file}" for file in files if file.lower().endswith('.svg')]
        svg_files = svg_files + svg_files_directory
    return svg_files

if __name__ == "__main__":
    # directory = input("Enter the directory path to search: ").strip()
    black_white_directory = './black/white'
    black_color = "rgb(27,27,29)"
    white_color = "rgb(255,255,255)"
    red_dark_color = "rgb(149,6,6)"
    red_burgundy_color = "rgb(102,0,51)"
    red_supreme_color = "rgb(237,29,36)"
    yellow_gold_color = "rgb(239,191,4)"
    green_dark_color = "rgb(6,64,43)"
    green_emerald_color = "rgb(0,103,79)"
    blue_dark_color = "rgb(17,17,132)"
    blue_navy_color = "rgb(0,0,128)"
    if os.path.isdir(black_white_directory):
        black_white_svg_files = get_svg_files(black_white_directory)
        print(f"Total number of .svg files: {len(black_white_svg_files)}")
        for black_white_svg_file in black_white_svg_files:
            black_svg_io = open(black_white_svg_file, 'r')
            black_svg_string = black_svg_io.read()
            black_count = black_svg_string.count(black_color)
            if black_count == 1:
                colored_svg_string = black_svg_string.replace(white_color, yellow_gold_color)
                colored_svg_file : str = black_white_svg_file.replace("black/white", "black/gold")
                colored_svg_path = Path("/".join(colored_svg_file.split("/")[:-1]))
                colored_svg_path.mkdir(parents=True, exist_ok=True)
                colored_svg_filename = Path(colored_svg_file)
                colored_svg_filename.touch(exist_ok=True)
                colored_svg_io = open(colored_svg_file, "w+")
                colored_svg_io.write(colored_svg_string)
                colored_svg_io.close()
                print(black_white_svg_file, black_count)
    else:
        print("The provided path is not a valid directory.")
```
